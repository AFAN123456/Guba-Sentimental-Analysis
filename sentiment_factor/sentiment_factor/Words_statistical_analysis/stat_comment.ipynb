```python
def unique_words(post_partition_dict, comment_partition_dict, limit = 5000):
    # To get a keywords list, which contains most comment words in comment and post

    tmp_list = []
    # find unique words in comments and posts
    for i in tqdm(post_partition_dict):
        for j in post_partition_dict[i]['words']:
            if j not in stopwords:
                tmp_list.append(j)
    for i in tqdm(comment_partition_dict):
        for j in comment_partition_dict[i]['words']:
            if j not in stopwords:
                tmp_list.append(j)

    freq_dic = collections.Counter(tmp_list)
    top_dic = freq_dic.most_common(limit)
    words_list = [i[0] for i in top_dic]
     
    return words_list
```

```python
def flatten(input_list):
    output_list = []
    while True:
        if input_list == []:
            break
        for index, i in enumerate(input_list):
            if type(i)== list:
                input_list = i + input_list[index+1:]
                break
            else:
                output_list.append(i)
                input_list.pop(index)
                break
    return output_list
```
