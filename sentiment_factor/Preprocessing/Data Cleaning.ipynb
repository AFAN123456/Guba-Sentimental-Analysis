```python
detail_url = detail[detail.detail_url.isna()==False]
```

```python
detail_url.detail_url
```

```python
def clean_url(row):
    temp = row.detail_url.split(',')[-1]
    temp = temp.split('.')[0]
    try:
        row['cleaned_url'] = int(temp)
    except ValueError:
        print(row._id)
    return row
cleaned_detail_url = detail_url.progress_apply(clean_url, axis = 1)
```

```python
detail_url[detail_url._id == '5f3402a155cacd9e45ea3321']
```

```python
cleaned_detail_url = cleaned_detail_url.drop(652798).reset_index(drop=True)
cleaned_detail_url['cleaned_url'] = cleaned_detail_url.cleaned_url.astype(int)
cleaned_detail_url = cleaned_detail_url.drop('_id', axis = 1)
```
